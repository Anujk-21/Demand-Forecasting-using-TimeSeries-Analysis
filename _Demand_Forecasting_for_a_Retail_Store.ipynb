```python
#train test split
train = df.loc[df.index < '12-01-2011']
test = df.loc[df.index >= '12-01-2011']

fig, ax = plt.subplots(figsize=(15,5))
train.plot(ax=ax, label='Training set')
test.plot(ax=ax, label='testing set')
```

```python
#feature selection
def create_feature(df):
  df['month'] = df.index.month
  df['dayofyear'] = df.index.dayofyear
  return df

df = create_feature(df)
print(df)
```

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error
```

```python
train = create_feature(train)
test = create_feature(test)
```

```python
df.columns
```

```python
features = ['Store', 'Weekly_Sales', 'Holiday_Flag', 'Temperature', 'Fuel_Price',
       'CPI', 'Unemployment', 'month', 'dayofyear']
target = ['Weekly_Sales']
```

```python
x_train = train[features]
y_train = train[target]

x_test = test[features]
y_test = test[target]
```

```python
reg = xgb.XGBRegressor(n_estimators = 1000, early_stopping_rounds = 50)
reg.fit(x_train,y_train,
        eval_set = [(x_train, y_train),(x_test, y_test)],
        verbose = 100)
```

```python
pd.DataFrame(reg.feature_importances_,
             index=reg.feature_names_in_,
             columns=['importance'])
```

```python
reg.predict(x_test) #predicted Values..
```

```python
test['prediction'] = reg.predict(x_test)
```

```python
df = df.merge(test[['prediction']], how='left',left_index = True, right_index = True)
df
```

```python
ax = df[['Weekly_Sales']].plot(figsize=(15,5))
df['prediction'].plot(ax=ax , style='.')
ax.set_title("OG data and Prediction")
plt.show()
```
